```python
# 櫨 1. Importando Bibliotecas
import pandas as pd

# 櫨 2. Carregando os Dados
# (Aqui vocﾃｪ importa do seu link Raw ou do arquivo local)
df = pd.read_csv("/content/dados_unificados.csv")

# 櫨 3. Excluindo a coluna 'registro_sistema'
if 'registro_sistema' in df.columns:
    df = df.drop(columns=['registro_sistema'])
else:
    print("Coluna 'registro_sistema' nﾃ｣o encontrada para exclusﾃ｣o.")

# 櫨 4. Convertendo a coluna 'matricula_ou_siape' para String
if 'matricula_ou_siape' in df.columns:
    df['matricula_ou_siape'] = df['matricula_ou_siape'].astype(str)
else:
    print("Coluna 'matricula_ou_siape' nﾃ｣o encontrada para conversﾃ｣o.")

# 櫨 5. Criando nova coluna 'classe_cdu' baseada na coluna 'localizacao'

def mapear_classe_cdu(localizacao):
    try:
        localizacao = int(str(localizacao)[:3])  # Garante pegar os 3 primeiros dﾃｭgitos
    except:
        return 'Desconhecido'

    if 0 <= localizacao <= 99:
        return 'Generalidades. Ciﾃｪncia e conhecimento'
    elif 100 <= localizacao <= 199:
        return 'Filosofia e psicologia'
    elif 200 <= localizacao <= 299:
        return 'Religiﾃ｣o'
    elif 300 <= localizacao <= 399:
        return 'Ciﾃｪncias sociais'
    elif 400 <= localizacao <= 499:
        return 'Classe vaga'
    elif 500 <= localizacao <= 599:
        return 'Matemﾃ｡tica e ciﾃｪncias naturais'
    elif 600 <= localizacao <= 699:
        return 'Ciﾃｪncias aplicadas'
    elif 700 <= localizacao <= 799:
        return 'Belas artes'
    elif 800 <= localizacao <= 899:
        return 'Linguagem. Lﾃｭngua. Linguﾃｭstica'
    elif 900 <= localizacao <= 999:
        return 'Geografia. Biografia. Histﾃｳria'
    else:
        return 'Fora da Classificaﾃｧﾃ｣o'

if 'localizacao' in df.columns:
    df['classe_cdu'] = df['localizacao'].apply(mapear_classe_cdu)
else:
    print("Coluna 'localizacao' nﾃ｣o encontrada para criaﾃｧﾃ｣o da classe CDU.")

# 櫨 6. Visualizando as alteraﾃｧﾃｵes
print(df.head())
```
